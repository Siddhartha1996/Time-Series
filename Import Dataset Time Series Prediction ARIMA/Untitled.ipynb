```python
df['Month'] = pd.to_datetime(df['Month'])
```

```python
df

#Here, converting to datetime, we automatically get DD as 01. And our format is YYYY-MM-DD
```

```python
data = df[["Month", "TotalAirportsWeight"]]
```

```python
data.set_index(df['Month'], inplace = True)
data = data.drop('Month',axis=1)
```

```python
plt.plot(data)
```

```python
from statsmodels.tsa.stattools import adfuller
adf = adfuller(data.TotalAirportsWeight)
print('p-value: ',adf[1])
```

```python
#Null Hypothesis is non stationarity, so here we have non stationary data
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(data.TotalAirportsWeight)
plot_pacf(data.TotalAirportsWeight)
```

```python
#Using First Differences because we have non stationarity. To try making it stationary, we use first differences:

plt.plot(data.TotalAirportsWeight.diff().dropna())
plot_acf(data.TotalAirportsWeight.diff().dropna())
plot_pacf(data.TotalAirportsWeight.diff().dropna());
```

```python
from  statsmodels.tsa.arima_model import ARIMA
arima = ARIMA(df.TotalAirportsWeight,order=(2,1,2))
model = arima.fit()
print(model.summary())
```

```python
model.plot_predict();
```

```python
#splitting the data into train and test sets
train = data[:130]
test = data[130:]
```

```python
arima = ARIMA(train,order = (2,1,2))
model = arima.fit()
print(model.summary())
```

```python
fcast,se,confidencebands = model.forecast(37,alpha=0.05)
```

```python
pred = pd.DataFrame(data=fcast,columns=['values'])
pred.set_index(test.index,inplace = True)
pred.tail(10)
```

```python
plt.figure(figsize = (20,10))
plt.plot(train,label = 'Trained')
plt.plot(test,label = 'Original')
plt.plot(pred, label = 'Predicted')
plt.legend
plt.xlabel('Month')
plt.ylabel('TotalAirportsWeight')
```
